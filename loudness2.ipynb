```python
import numpy as np
import librosa
```

```python
Fs = 22050
x, Fs = librosa.load('data/lofi-type-beat.mp3', sr=Fs, mono=True)
```

```python
rms = librosa.feature.rms(y=x)[0]
loudness = (rms - np.min(rms)) / (np.max(rms) - np.min(rms))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(loudness)
```
